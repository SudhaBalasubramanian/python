<!-- poem.txt contains famous poem "Road not taken" by poet Robert Frost.
 You have to read this file in your python program and find out words with maximum Occurrence. -->   

```python
word_stats = {}
with open("poem.txt", 'r') as f:
    for line in f:
        words = line.replace(',', '').replace('\n', '').split(' ')
        for word in words:
            if word in word_stats:
                word_stats[word] +=1
            else:
                word_stats[word] = 1
print(word_stats)

max_value = max(word_stats.values())

max_keys = [k for k,v in word_stats.items() if v==max_value]
print(max_keys)
print(max_value)
```

<!-- stocks.csv contains stock price, earnings per share and book value.
 You are writing a stock market application that will process this file and create a new file with financial metrics such as pe ratio and price to book ratio. These are calculated as,
pe ratio = price / earnings per share
price to book ratio = price / book value -->

<!-- Your input format (stocks.csv) is,

Company Name	Price	Earnings Per Share	Book Value
Reliance	1467	66	653
Tata Steel	391	89	572
Output.csv should look like this,

Company Name	PE Ratio	PB Ratio
Reliance	22.23	2.25
Tata Steel	4.39	0.68 -->

```python
with open('stocks.csv', 'r') as f, open('stocks_metrics.csv', 'w') as out:
    out.write('Company Name, PE Ratio, PB Ratio\n')
    next(f)
    for line in f:
        words = line.split(',')
        company_name = words[0]
        pe_ratio = round(float(words[1])/float(words[2]), 2)
        pb_ratio = round(float(words[1])/float(words[3]), 2)
        out.write(f'{company_name}, {pe_ratio}, {pb_ratio}\n')
```
